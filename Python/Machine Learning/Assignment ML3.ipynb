```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/daniel-dc-cd/data_science/master/module_4_ML/data/seattle_weather_1948-2017.csv')
```

```python
numrows = 25549
```

```python
heuristic_df = pd.DataFrame({'yesterday':[0.0]*numrows,
                             'today':[0.0]*numrows,
                             'tomorrow':[0.0]*numrows,
                             'guess':[False]*numrows, 
                             'rain_tomorrow':[False]*numrows, 
                             'correct':[False]*numrows, 
                             'true_positive':[False]*numrows, 
                             'false_positive':[False]*numrows,
                             'true_negative':[False]*numrows, 
                             'false_negative':[False]*numrows}) 
```

```python
seq = ['yesterday',
       'today',
       'tomorrow',
       'guess',
       'rain_tomorrow',
       'correct',
       'true_positive',
       'false_positive',
       'true_negative',
       'false_negative']
heuristic_df = heuristic_df.reindex(columns=seq)
```

```python
for z in range(numrows):
  
    i = z + 2
    
    yesterday = df.iloc[(i-2),1]
    today = df.iloc[(i-1),1]
    tomorrow = df.iloc[i,1]
    rain_tomorrow = df.iloc[(i),1]
    
    heuristic_df.iat[z,0] = yesterday
    heuristic_df.iat[z,1] = today
    heuristic_df.iat[z,2] = tomorrow
    heuristic_df.iat[z,3] = False 
    heuristic_df.iat[z,4] = rain_tomorrow
    
   
    if today > 0.0 and yesterday > 0.0:
        heuristic_df.iat[z,3] = True
    if heuristic_df.iat[z,3] == heuristic_df.iat[z,4]:
        heuristic_df.iat[z,5] = True
        if heuristic_df.iat[z,3] == True:
            heuristic_df.iat[z,6] = True 
        else:
            heuristic_df.iat[z,8] = True 
    else:
        heuristic_df.iat[z,5] = False
        if heuristic_df.iat[z,3] == True:
            heuristic_df.iat[z,7] = True 
        else:
            heuristic_df.iat[z,9] = True 
```

```python
from sklearn.model_selection import train_test_split 
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

h_train , h_test , h_trainy , h_testy =train_test_split(heuristic_df ,heuristic_df['rain_tomorrow'] ,test_size=.8)
h_train.head()
```

```python
from sklearn.metrics import mean_squared_error
def Calucations(NTP_TRIAN , NTP_TEST , NFP_TRIAN , NFP_TEST , NFN_train , NFN_test , true , pred):
    #percent of the time we where correct
    #(true positive + true negative)/ number of guesses
    prect_train = NTP_TRIAN/(NTP_TRIAN+NFP_TRIAN)
    prect_test = NTP_TEST/(NTP_TEST+NFP_TEST)
    
    # recall the percent 
    # (num true positive)/(num tru positive + num false negative)
    Rec_train=NTP_TRIAN/(NTP_TRIAN+NFN_train)
    Rec_test=NTP_TEST/(NTP_TEST+NFN_test)
    
    squared_errors = (true - pred) ** 2
    SSE=np.sum(squared_errors) 
    
    return print("Training Precision:" , prect_train*100,"%\n" , "Testing Precision:" , prect_test , "%\n\n",
              "Training Recall:" ,  Rec_train*100 , "%\n" , "Testing Recall:" , Rec_test*100 , "%\n\n" , "the SSE" ,
               SSE )
```
